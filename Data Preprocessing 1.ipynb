```python
df.drop(['Unnamed: 0', 'Unique Key','Location Type','Created Date','Closed Date','Incident Address',
         'Street Name','Address Type','City',
         'Status', 'Resolution Description', 'Latitude', 'Longitude'],inplace=True, axis=1)
```

```python
df1 = df.copy()
```

```python
df1
```

```python
complaints = df1.groupby(['Complaint Type'], as_index=False).count()
```

```python
complaints
```

```python
complaints.plot(x='Complaint Type',kind='bar', figsize=(20,15))
plt.show()
```

```python
combyborough = df1.groupby(['Complaint Type','Borough'], as_index=False).count()
```

```python
combyborough
```

```python
combyborough.plot(x='Borough',kind='bar', figsize=(20,15))
plt.show()
```
